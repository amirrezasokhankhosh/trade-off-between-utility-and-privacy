```python
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("./pr_synthetic_results/syn.csv")
df.head()
```

```python
y = tf.constant(np.array(df[["salary_class"]].copy()))
X = tf.constant(np.array(df.drop(["salary_class"], axis=1)))
```

```python
X_train, y_train = X[:-4000], y[:-4000]
X_val, y_val = X[-4000:-2000], y[-4000:-2000]
X_test, y_test = X[-2000:], y[-2000:]
```

```python
W = tf.Variable(np.random.rand(X_train.shape[1], 1)*0.1)
b = tf.Variable(0, dtype=tf.double)
```

```python
learning_rate = 3e-2
epochs = 1000
```

```python
y_train.shape
```

```python
history = []
for _ in range(epochs):
    with tf.GradientTape() as tape:
        tape.watch([W, b])
        pred = tf.sigmoid(X_train @ W + b)
        loss = tf.losses.binary_crossentropy(y_train, pred, axis=0)
    pred_val = tf.sigmoid(X_val @ W + b)
    loss_val = tf.losses.binary_crossentropy(y_val, pred_val, axis=0)

    print(loss.numpy()[0], loss_val.numpy()[0])

    history.append(loss.numpy()[0])

    grad = tape.gradient(loss, {
        "W" : W,
        "b" : b
    })

    W = W - learning_rate * grad["W"]
    b = b - learning_rate * grad["b"]
```
